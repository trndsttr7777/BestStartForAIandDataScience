# import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas import plotting

import seaborn as sns

# 加载 iris 数据集
df = sns.load_dataset('iris')
```

# 数据可视化
encode之后会显示class，但是没必要

```python
sns.lineplot(df)
```

```python
sns.pointplot(df)
```

```python
sns.kdeplot(df)

plt.show()# plt.show()用于在窗口中显示图像，但是在.ipynb的默认环境中，可以不需要
```

```python
sns.violinplot(df)
plt.show()
```

### 因为种类数量对于violinplot是无效的，所以

```python
sns.violinplot(df[df.columns[:-1]])
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue= 'species') # 默认pairplot()会用多种颜色标记出来
# sns.pairplot(data=iris, palette=antV, hue= 'species')
```

```python
fig=plt.gcf()
fig = plt.gcf()  # 获取当前图形对象（Get Current Figure），用于后续对图形进行操作

# 计算仅针对数值特征的相关性矩阵
numerical_iris = df.select_dtypes(include=np.number)  # 选择数据集中仅包含数值类型的列

# 使用 seaborn 的 heatmap 函数绘制热力图
fig = sns.heatmap(
    numerical_iris.corr(),  # 计算数值列的相关性矩阵
    linewidths=1,           # 设置每个单元格之间的线宽为1
    vmin=-1,                # 设置颜色映射的最小值为-1（完全负相关）
    vmax=1,                 # 设置颜色映射的最大值为1（完全正相关）
)

plt.show()  # 显示图形
```
