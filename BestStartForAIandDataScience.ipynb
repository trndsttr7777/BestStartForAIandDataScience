### violinplot()

```python
# 设置颜色主题
antV = ['#1890FF', '#2FC25B', '#FACC14']
# 绘制 Violinplot 小钢琴图
f, axes = plt.subplots(2, 2, figsize=(8, 8), sharex=True)
sns.despine(left=True)
# ax参数是用来设定同时展示多张图片的

# Changed column names to match the actual column names in the iris dataset
sns.violinplot(x='species', y='sepal_length', data=iris, palette=antV, ax=axes[0, 0])
sns.violinplot(x='species', y='sepal_width', data=iris, palette=antV, ax=axes[0, 1])
sns.violinplot(x='species', y='petal_length', data=iris, palette=antV, ax=axes[1, 0])
sns.violinplot(x='species', y='petal_width', data=iris, palette=antV, ax=axes[1, 1])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加载 iris 数据集
iris = sns.load_dataset('iris')

# 使用 implot 绘制散点图和回归线
sns.lmplot(
    x='sepal_length',  # x 轴数据列
    y='sepal_width',   # y 轴数据列
    data=iris,         # 数据集
    hue='species',     # 根据 species 列分组，不同组用不同颜色
    col='species',     # 根据 species 列分面，每个 species 一个子图
    palette='Set1',    # 设置调色板
    col_wrap=2,        # 每行显示 2 个子图
    height=4,          # 每个子图的高度
    aspect=1,          # 每个子图的宽高比
    markers=['o', 's', 'D'],  # 不同组的标记样式
    sharex=True,       # 共享 x 轴
    sharey=True,       # 共享 y 轴
    legend=True,       # 显示图例
    order=1,           # 使用一阶多项式回归
    logistic=False,    # 不使用逻辑回归
    lowess=False,      # 不使用局部加权回归
    robust=False,      # 不使用鲁棒回归
    logx=False,        # 不对 x 轴取对数
    x_estimator=None,  # 不对 x 值进行估计
    x_bins=None,       # 不分组 x 值
    x_ci=95,           # 设置 x 值的置信区间为 95%
    scatter=True,      # 绘制散点图
    fit_reg=True,      # 拟合回归线
    ci=95,             # 设置回归线的置信区间为 95%
    n_boot=1000,       # 使用 1000 个 bootstrap 样本
    units=None,        # 不使用多级 bootstrap
    seed=None,         # 不设置随机种子
    scatter_kws={'s': 50},  # 设置散点图的大小
    line_kws={'lw': 2},    # 设置回归线的线宽
    facet_kws={'despine': False}  # 不移除子图的边框
)

# 显示图形
plt.show()
```
